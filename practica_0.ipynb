## Objetivo

## Introducción

ROS2 (Robot Operating System 2) es un framework utilizado en robótica para facilitar la comunicación entre diferentes procesos. Estos procesos se organizan en nodos, los cuales pueden intercambiar información a través de tópicos.

En esta práctica se trabajó dentro de una máquina virtual con Linux, usando la terminal, Visual Studio Code y ROS2 Jazzy. Se creó y compiló un paquete en Python, dentro del cual se desarrollaron dos nodos: un nodo publicador y un nodo suscriptor. El propósito fue observar cómo un nodo puede enviar información periódicamente y cómo otro nodo puede recibirla y procesarla.

## Marco teórico

### ¿Qué es un nodo?
Un nodo es un proceso dentro de ROS2 que realiza una tarea específica. Por ejemplo, un nodo puede leer sensores, publicar datos, recibir información o controlar un actuador.

### ¿Qué es un tópico?
Un tópico es un canal de comunicación por el cual los nodos intercambian mensajes. Un nodo puede publicar información en un tópico y otro nodo puede suscribirse a ese mismo tópico para recibirla.

### Modelo publicador-suscriptor
En ROS2, el modelo publicador-suscriptor permite desacoplar la comunicación entre nodos. El nodo publicador envía mensajes a un tópico, mientras que el nodo suscriptor recibe esos mensajes y puede utilizarlos para ejecutar alguna acción o procesamiento.

## Desarrollo

Para realizar la práctica, primero se localizó el workspace de ROS2 dentro de la carpeta `ws_sem26_2`. Posteriormente, se identificó el paquete en Python llamado `p0_py`, el cual contenía los archivos principales del ejercicio.

Se revisó y corrigió la configuración del paquete, especialmente el archivo `setup.py`, para registrar correctamente los nodos y poder ejecutarlos mediante el comando `ros2 run`.

Después, se revisaron y corrigieron los archivos de Python correspondientes a los nodos:
- `primer_nodo.py`
- `subscriber_node.py`

Durante este proceso se corrigieron errores de sintaxis, indentación, nombres de funciones especiales como `__init__`, y la función `main`, para que ROS2 pudiera reconocer y ejecutar correctamente ambos nodos.

Una vez corregidos los archivos, se compiló el workspace con:

`colcon build --symlink-install`

Después se cargó el entorno con:

`source install/setup.bash`

Finalmente, se ejecutaron los nodos en terminales separadas para verificar la comunicación.

## Descripción del nodo publicador

El nodo publicador fue implementado en el archivo `primer_nodo.py`. Su función es generar y enviar mensajes de manera periódica a través de un tópico.

En la primera versión del ejercicio, el nodo publicaba datos enteros de tipo `Int64`. Posteriormente, para acercarse mejor a lo solicitado en la práctica, se ajustó el enfoque del ejercicio para trabajar con el modelo publicador-suscriptor y verificar el intercambio de mensajes entre nodos.

El publicador utiliza un temporizador para ejecutar su callback cada cierto tiempo. Cada vez que se activa, construye un mensaje y lo publica en el tópico correspondiente.

## Descripción del nodo suscriptor

El nodo suscriptor fue implementado en el archivo `subscriber_node.py`. Este nodo se suscribe al mismo tópico utilizado por el nodo publicador, de manera que puede recibir los mensajes emitidos por éste.

Cuando el nodo suscriptor recibe un mensaje, ejecuta un callback para procesar la información. En la práctica, este nodo permitió comprobar que la comunicación entre nodos se estaba realizando correctamente.

Además, se pudo observar en consola la activación del nodo suscriptor y su funcionamiento dentro del sistema ROS2.

## Dificultades encontradas

Durante la práctica se presentaron varias dificultades, principalmente relacionadas con la estructura del paquete y con errores de sintaxis en Python.

Entre los problemas encontrados estuvieron:
- confusión entre el nombre del paquete y el nombre de la carpeta interna,
- errores al registrar los nodos en `setup.py`,
- problemas con la indentación en Python,
- uso incorrecto de `__init__`,
- errores en la definición de la función `main`,
- ejecución de `colcon build` fuera del workspace correcto.

La corrección de estos detalles permitió comprender mejor cómo se organiza un proyecto en ROS2 y la importancia de seguir una estructura precisa.
